Q2. Perform the Exploratory Data Analysis on your domain-based dataset and demonstrate
the retrieved insights using “Matplotlib” modules. Visualize hidden insights using appropriate
plots (graphs) [Usage of line plot and scatter plot are mandatory]

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

csv_file = 'Video_Games_Sales_as_at_22_Dec_2016.csv'
df = pd.read_csv(csv_file)

# Line Graph
plt.figure(figsize=(12, 6))
plt.plot(df['Year_of_Release'], df['Global_Sales'], marker='o')
plt.title('Line Graph')
plt.xlabel('Year_of_Release')
plt.ylabel('Global_Sales')
plt.grid(True)




plt.show()
```

```python

# Bar Graph
plt.figure(figsize=(10, 6))
plt.bar(df['Year_of_Release'], df['Global_Sales'])
plt.title('Bar Graph')
plt.xlabel('Year_of_Release')
plt.ylabel('Global_Sales')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Critic_Count'], df['Critic_Score'], c='b', marker='o')
plt.title('Scatter Plot')
plt.xlabel('Critic_Count')
plt.ylabel('Critic_Score')
plt.grid(True)
plt.show()
```
